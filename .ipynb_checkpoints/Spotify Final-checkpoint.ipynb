# Joe Dominik's Making A Harmonic DJ Set - Producing the Algorithm

This notebook explains the two different algorithms that were made to produce a harmonic DJ set. The first being inefficient compared to the second algorithm.

## Importing all needed libraries for making the algorithm

Note: To show a significant difference in efficiency between the two algorithms, a playlist of 10,000 songs was chosen from Spotify.

```python
#spotipy for Spotify API
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
#matplotlib.pyplot and seaborn for EDA
import matplotlib.pyplot as plt
import seaborn as sns
#numpy and pandas for dataframe needs
import pandas as pd
import random
import copy
import time
#client credentials
client_id = "insert here"
client_secret="insert here"
#10,000 song playlist on Spotify (Alex Marty)
longest_playlist_id = 'spotify:playlist:6FKDzNYZ8IW1pvYVF4zUN2'
user = 'a6b7aa5eda3847c0'
```

The Camelot Wheel is needed for each song to produce a harmonic DJ set. Spotify provides *key* and *mode*. Using the dictionary from https://maustsontoast.com/2020/pitch-class-tonal-counterparts-and-camelot-key-equivalents the equivalent Camelot "Score" was added to each song.

If you are unsure for the reasoning behind using the Camelot Score, please refer to my ReadMe.

```python
camelot_scores = {
    (0,1):'8B',
    (1,1):'3B',
    (2,1):'10B',
    (3,1):'5B',
    (4,1):'12B',
    (5,1):'7B',
    (6,1):'2B',
    (7,1):'9B',
    (8,1):'4B',
    (9,1):'11B',
    (10,1):'6B',
    (11,1):'1B',
    (0,0):'5A',
    (1,0):'12A',
    (2,0):'7A',
    (3,0):'2A',
    (4,0):'9A',
    (5,0):'4A',
    (6,0):'11A',
    (7,0):'6A',
    (8,0):'1A',
    (9,0):'8A',
    (10,0):'3A',
    (11,0):'10A',
}
def camelot_function(key, mode):
    return camelot_scores.get((key, mode))
```

## The creating of the data frame and dictionary that will be used in the algorithms

```python
#using a playlist of 10,000 songs
#using credentials from first cell
spotify = spotipy.Spotify(client_credentials_manager = SpotifyClientCredentials(client_id = client_id,
                                                           client_secret = client_secret))
inefficient_list = []
efficient_list = []
#getting the exact playlist from the user
results = spotify.user_playlist_tracks(user = user, playlist_id = longest_playlist_id)
#getting the items from the playlist, getting rid of metadata that isn't needed
tracks = results['items']
#using same while loop until it is at the end of the playlist
while results['next']:
    results = spotify.next(results)
    tracks.extend(results['items'])

longest_list = []
longest_dict = {
    '1A': [],
    '2A': [],
    '3A': [],
    '4A': [],
    '5A': [],
    '6A': [],
    '7A': [],
    '8A': [],
    '9A': [],
    '10A': [],
    '11A': [],
    '12A': [],
    '1B': [],
    '2B': [],
    '3B': [],
    '4B': [],
    '5B': [],
    '6B': [],
    '7B': [],
    '8B': [],
    '9B': [],
    '10B': [],
    '11B': [],
    '12B': [],
}
for track in tracks:
    #getting the audio features for each track
    audio_features = spotify.audio_features(track['track']['id'])
    #appending to the list the track name, artist, popularity, tempo, key, danceability, energy, and loudness
    longest_list.append([track['track']['name'], track['track']['artists'][0]['name'], track['track']['popularity'], 
                       audio_features[0]['tempo'], audio_features[0]['key'], audio_features[0]['mode'], 
                       audio_features[0]['danceability'], 
                       audio_features[0]['energy'], audio_features[0]['loudness'], 
                       camelot_function(audio_features[0]['key'], audio_features[0]['mode'])])
    longest_dict[camelot_function(audio_features[0]['key'], audio_features[0]['mode'])].append([track['track']['name'],
                        track['track']['artists'][0]['name'], track['track']['popularity'], 
                       audio_features[0]['tempo'], audio_features[0]['key'], audio_features[0]['mode'], 
                       audio_features[0]['danceability'], 
                       audio_features[0]['energy'], audio_features[0]['loudness'], 
                       camelot_function(audio_features[0]['key'], audio_features[0]['mode'])])
#making the dataframe after done with the for loop
longest_df = pd.DataFrame(longest_list, columns = ['Song Name', 'Artist', 'Popularity', 'Tempo', 'Key', 'Mode',
                                               'Danceability', 'Energy', 'Loudness', 'Camelot Score'])
```

This first algorithm uses several if statements to match the previous song to the next song that is within the correct Camelot Score.

```python
#function is used to match songs with similar keys with tempos of +/- 5
def harmony_function(prev_song, next_song, tempo1, tempo2):
    if(abs(tempo2 - tempo1) < 5):
        if(prev_song == "1A"):
            if((next_song == "12A") or (next_song == "1A") 
               or (next_song == "2A") or (next_song == "1B")):
                return True
            else: 
                return False
        elif (prev_song == "1B"):
            if((next_song == "12B") or (next_song == "1B") 
               or (next_song == "2B") or (next_song == "1A")):
                return True
            else: 
                return False
        elif (prev_song == "2A"):
            if((next_song == "1A") or (next_song == "2A") 
               or (next_song == "3A") or (next_song == "2B")):
                return True
            else: 
                return False
        elif (prev_song == "2B"):
            if((next_song == "1B") or (next_song == "2B") 
               or (next_song == "3B") or (next_song == "2A")):
                return True
            else: 
                return False
        elif (prev_song == "3A"):
            if((next_song == "2A") or (next_song == "3A") 
               or (next_song == "4A") or (next_song == "3B")):
                return True
            else: 
                return False
        elif (prev_song == "3B"):
            if((next_song == "2B") or (next_song == "3B") 
               or (next_song == "4B") or (next_song == "3A")):
                return True
            else: 
                return False
        elif (prev_song == "4A"):
            if((next_song == "3A") or (next_song == "4A") 
               or (next_song == "5A") or (next_song == "4B")):
                return True
            else: 
                return False
        elif (prev_song == "4B"):
            if((next_song == "3B") or (next_song == "4B") 
               or (next_song == "5B") or (next_song == "4A")):
                return True
            else: 
                return False
        elif (prev_song == "5A"):
            if((next_song == "4A") or (next_song == "5A") 
               or (next_song == "6A") or (next_song == "5B")):
                return True
            else: 
                return False
        elif (prev_song == "5B"):
            if((next_song == "4B") or (next_song == "5B") 
               or (next_song == "6B") or (next_song == "5A")):
                return True
            else: 
                return False
        elif (prev_song == "6A"):
            if((next_song == "5A") or (next_song == "6A") 
               or (next_song == "7A") or (next_song == "6B")):
                return True
            else: 
                return False
        elif (prev_song == "6B"):
            if((next_song == "5B") or (next_song == "6B") 
               or (next_song == "7B") or (next_song == "6A")):
                return True
            else: 
                return False
        elif (prev_song == "7A"):
            if((next_song == "6A") or (next_song == "7A") 
               or (next_song == "8A") or (next_song == "7B")):
                return True
            else: 
                return False
        elif (prev_song == "7B"):
            if((next_song == "6B") or (next_song == "7B") 
               or (next_song == "8B") or (next_song == "7A")):
                return True
            else: 
                return False
        elif (prev_song == "8A"):
            if((next_song == "7A") or (next_song == "8A") 
               or (next_song == "9A") or (next_song == "8B")):
                return True
            else: 
                return False
        elif (prev_song == "8B"):
            if((next_song == "7B") or (next_song == "8B") 
               or (next_song == "9B") or (next_song == "8A")):
                return True
            else: 
                return False
        elif (prev_song == "9A"):
            if((next_song == "8A") or (next_song == "9A") 
               or (next_song == "10A") or (next_song == "9B")):
                return True
            else: 
                return False
        elif (prev_song == "9B"):
            if((next_song == "8B") or (next_song == "9B") 
               or (next_song == "10B") or (next_song == "9A")):
                return True
            else: 
                return False
        elif (prev_song == "10A"):
            if((next_song == "9A") or (next_song == "10A") 
               or (next_song == "11A") or (next_song == "10B")):
                return True
            else: 
                return False
        elif (prev_song == "10B"):
            if((next_song == "9B") or (next_song == "10B") 
               or (next_song == "11B") or (next_song == "10A")):
                return True
            else: 
                return False
        elif (prev_song == "11A"):
            if((next_song == "10A") or (next_song == "11A") 
               or (next_song == "12A") or (next_song == "11B")):
                return True
            else: 
                return False
        elif (prev_song == "11B"):
            if((next_song == "10B") or (next_song == "11B") 
               or (next_song == "12B") or (next_song == "11A")):
                return True
            else: 
                return False
        elif (prev_song == "12A"):
            if((next_song == "11A") or (next_song == "12A") 
               or (next_song == "1A") or (next_song == "12B")):
                return True
            else: 
                return False
        elif (prev_song == "12B"):
            if((next_song == "11B") or (next_song == "12B") 
               or (next_song == "1B") or (next_song == "12A")):
                return True
            else: 
                return False
    else:
        return False
```

## Using the first algorithm:

Note: When running both algorithms, the song that was indexed at 5000 was used each time. This was kept consistent to help decide which algorithm was more efficient.

```python
longest_list_2 = []
#deep copying the dataframe of the playlist so I can rerun the program multiple times without having 
#to reconstruct the dataframe
longest_df_2 = longest_df.copy(deep = True)
#starting the confirmation for the first song with N
confirmation = "N"
print("This is a tool to help you build a harmonic set from your desired playlist.")
#this while loop goes until something other than N is put in
while(confirmation == "N"):
    #asks the user for a number between the starting and ending index of the playlist
    starting_song_index = input("To start the set please enter a number between " + str(0) + " and " 
                                + str(len(longest_df_2) - 1) + " : ")
    starting_song_index = int(starting_song_index)
    #finds the song in the data frame and asks the user if this is the correct song
    starting_song = longest_df_2.iloc[starting_song_index]
    print("Starting song for set: " + str(starting_song['Song Name']) + " by " + str(starting_song['Artist']))
    confirmation = input("Is this correct? Type Y for yes and N for no: ")
#after finding the correct song, the tempo and camelot score of that song is found
start = time.perf_counter()
prev_song_camelot = starting_song['Camelot Score']
prev_song_tempo = starting_song['Tempo']
#dropping this song from the dataframe
longest_df_2 = longest_df_2.drop(starting_song_index, axis = 0)
#reindexing to help with while loop
longest_df_2.index = range(len(longest_df_2))
starting_song = starting_song.tolist()
longest_list_2.append(starting_song)
#defining i as a counter if it cannot find a song on the first try
i = 0
while((len(longest_df_2) > 0) and (i < 5000)):
    #using the random portion of the algorithm to pick the next song
    temp = random.randint(0,len(longest_df_2) - 1)
    song = longest_df_2.iloc[temp]
    #finding the next song's camelot score and tempo
    next_song_camelot = song['Camelot Score']
    next_song_tempo = song['Tempo']
    #using the harmony function constructed above to evaluate if it is a match
    if(harmony_function(prev_song_camelot, next_song_camelot, prev_song_tempo, next_song_tempo)):
        longest_df_2 = longest_df_2.drop(temp, axis = 0)
        song_copy = song.copy(deep = True)
        song_copy = song_copy.tolist()
        longest_list_2.append(song_copy)
        prev_song_camelot = next_song_camelot
        prev_song_tempo = next_song_tempo
        i = 0
    else:
        #if it is not, then another iteration of the while loop happens
        i = i + 1
    longest_df_2.index = range(len(longest_df_2))
end = time.perf_counter()
performance = end - start
#the final playlist dataframe is constructed to make it easier to read
longest_df_3 = pd.DataFrame(longest_list_2, columns = ['Song Name', 'Artist', 'Popularity', 'Tempo', 'Key', 'Mode', 
                                                           'Danceability', 'Energy', 'Loudness', 'Camelot Score'])  
inefficient_list.append([len(longest_list_2), performance])
```

## Creating the dictionary of Camelot Scores that are harmonic to each other for the second algorithm: 

```python
harmony_dict = {
    '1A':['12A', '1A', '2A', '1B'],
    '2A':['1A', '2A', '3A', '2B'],
    '3A':['2A', '3A', '4A', '3B'],
    '4A':['3A', '4A', '5A', '4B'],
    '5A':['4A', '5A', '6A', '5B'],
    '6A':['5A', '6A', '7A', '6B'],
    '7A':['6A', '7A', '8A', '7B'],
    '8A':['7A', '8A', '9A', '8B'],
    '9A':['8A', '9A', '10A', '9B'],
    '10A':['9A', '10A', '11A', '10B'],
    '11A':['10A', '11A', '12A', '11B'],
    '12A':['11A', '12A', '1A', '12B'],
    '1B':['12B', '1B', '2B', '1A'],
    '2B':['1B', '2B', '3B', '2A'],
    '3B':['2B', '3B', '4B', '3A'],
    '4B':['3B', '4B', '5B', '4A'],
    '5B':['4B', '5B', '6B', '5A'],
    '6B':['5B', '6B', '7B', '6A'],
    '7B':['6B', '7B', '8B', '7A'],
    '8B':['7B', '8B', '9B', '8A'],
    '9B':['8B', '9B', '10B', '9A'],
    '10B':['9B', '10B', '11B', '10A'],
    '11B':['10B', '11B', '12B', '11A'],
    '12B':['11B', '12B', '1B', '12A'],
}
def harmony_function_2(score):
    return harmony_dict.get(score)
```

## Using the second algorithm: 

```python
longest_dict_2 = copy.deepcopy(longest_dict) 
final_playlist_list = []
#constructing the list for the songs that satisfy the constraints later in the program
potential_song_list = []
#starting the confirmation for the first song with N
confirmation = "N"
print("This is a tool to help you build a harmonic set from your desired playlist.")
#this while loop goes until something other than N is put in
while(confirmation == "N"):
    #asks the user for a number between the starting and ending index of the playlist
    starting_song_index = input("To start the set please enter a number between " + str(0) +
                                " and " + str(len(longest_df) - 1) + " : ")
    starting_song_index = int(starting_song_index)
    #finds the song in the data frame and asks the user if this is the correct song
    starting_song = longest_df.iloc[starting_song_index]
    print("Starting song for set: " + str(starting_song['Song Name']) + " by " + str(starting_song['Artist']))
    confirmation = input("Is this correct? Type Y for yes and N for no: ")
#after finding the correct song, the tempo and camelot score of that song is found
start = time.perf_counter()
previous_song_camelot = starting_song['Camelot Score']
previous_song_tempo = starting_song['Tempo']
starting_song = starting_song.tolist()
#the song is then removed from the value from the specific camelot score it was in
value_from_specific_camelot = longest_dict_2.pop(previous_song_camelot)
value_from_specific_camelot.remove(starting_song)
#the other songs that were not affected were put back into the dictionary
longest_dict_2[previous_song_camelot] = value_from_specific_camelot
#the first song is then added to the final playlist
final_playlist_list.append(starting_song)
indicator = True
#this while loop goes through the rest of the dictionary, adding to the final playlist if the constraints are met
while((len(final_playlist_list) < len(longest_df)) and (indicator == True)):
    #this finds all songs that are within the 4 camelot scores of the previous song
    corresponding_camelot_scores = harmony_function_2(previous_song_camelot)
    for camelot_score in corresponding_camelot_scores:
        songs = longest_dict_2[camelot_score]
        for song in songs:
            #this constraint makes sure no songs outside of 5 BPM are added to the list of potential songs
            if(abs(song[3] - previous_song_tempo) < 5):
                potential_song_list.append(song)
    #making sure the potential song list is not empty
    if(len(potential_song_list) > 0 ):
        #in order to keep randomness with the playlist, a random integer is generated
        random_int = random.randint(0,len(potential_song_list) - 1)
        #this next song is then found, the camelot score and tempo for that song are found
        next_song = potential_song_list[random_int]
        next_song_camelot = next_song[9]
        next_song_tempo = next_song[3]
        #the next song is then removed from the dictionary at the specific camelot score
        value_from_specific_camelot = longest_dict_2.pop(next_song_camelot)
        value_from_specific_camelot.remove(next_song)
        longest_dict_2[next_song_camelot] = value_from_specific_camelot
        #finally, the final playlist is added with the next song
        final_playlist_list.append(next_song)
        #the previous song camelot score and tempo are the next song camelot score and tempo, respectfully
        previous_song_camelot = next_song_camelot
        previous_song_tempo = next_song_tempo
        #to keep the potential song list empty, it is cleared
        potential_song_list.clear()
    else:
        indicator = False
end = time.perf_counter()
performance = end - start
#the final playlist dataframe is constructed to make it easier to read
playlist_df = pd.DataFrame(final_playlist_list, columns = ['Song Name', 'Artist', 'Popularity', 'Tempo', 'Key', 'Mode', 
                                                           'Danceability', 'Energy', 'Loudness', 'Camelot Score'])

efficient_list.append([len(final_playlist_list),performance])
```

## Data Analysis on the different algorithms

```python
inefficient_df = pd.DataFrame(inefficient_list, columns = ['Songs', 'Seconds'])
efficient_df = pd.DataFrame(efficient_list, columns = ['Songs', 'Seconds'])
print("After running the ineffienct and effient codes 100 times for playlist making: ")
print("Inefficient code: ")
print("Average songs in each set: " + str(inefficient_df['Songs'].mean()))
print("Average seconds to make each set: " + str(inefficient_df['Seconds'].mean()))
print("Efficient code: ")
print("Average songs in each set: " + str(efficient_df['Songs'].mean()))
print("Average seconds to make each set: " + str(efficient_df['Seconds'].mean()))
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(inefficient_df['Seconds'], label = "Inefficient")
plt.plot(efficient_df['Seconds'], label = "Efficient")
plt.xlabel('Code Completions', size=16)
plt.ylabel('Seconds to Complete Code', size=16)
plt.legend()
plt.show()
plt.figure(figsize=(10,5))
sns.distplot(inefficient_df['Seconds'], bins = 30)
plt.figure(figsize=(10,5))
sns.distplot(efficient_df['Seconds'], bins = 30)
```

After running each algorithm 100 times, the second algorithm, using dictionaries instead of multiple if statements, is more efficient. Not only is it, on average, 48 times faster to produce a harmonic set, but it also gives 15.8% more songs in the set than the first algorithm. This is the algorithm that will be used to make harmonic sets from the playlists I have created in Spotify.
